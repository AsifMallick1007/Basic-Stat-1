# Discriptive Statistics

```python
#Data
age_batch1 = [15,16,20,21,21,
21,21,21,21,21,
21,21,21,21,21,
21,21,21,22,22,
22,22,22,22,22,
23,23,23,23,23,
24,24,24,26,26,
27,27,28,35,56]
```

```python
import pandas as pd

data = pd.DataFrame(age_batch1)
data
```

```python
mean = data.mean()
median = data.median()
mode = data.mode()
var = data.var()
SD = data.std()
sk = data.skew()
kur = data.kurt()
kurt = data.kurtosis()
```

```python
print("The mean age is",mean)
print("The Median age is",median)
print("The Mode age is",mode)
print("The varince age is",var)
print("The S.D age is",SD)
print("The skweness of age is",sk)
print("The skweness of age is",kur)
print("The skweness of age is",kurt)
```
